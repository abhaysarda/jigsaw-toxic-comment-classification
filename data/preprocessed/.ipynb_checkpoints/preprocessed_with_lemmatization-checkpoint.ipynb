## Import Modules

```python
import numpy as np 
import pandas as pd 
import os
import re
import string
from string import digits
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

import warnings
warnings.filterwarnings('ignore')
```

## Import Data 

```python
train_data = pd.read_csv('../raw_data/train.csv')
test_data = pd.read_csv('../raw_data/test.csv')
```

```python
train_data.drop(["Unnamed: 0"], axis =1)
test_data.drop(["Unnamed: 0"], axis =1)
```

```python
train_comments = train_data["Comment"]
test_comments = test_data["Comment"]
```

```python
train_comments.head(25)
```

## Removing Punctuation

Primary functions used are str.translate() with str.maketrans() that is similar to a reference table which helps the translate method to remove punctuation. 

References: 

[1] https://www.tutorialspoint.com/python3/string_translate.htm

[2] https://stackoverflow.com/questions/34293875/how-to-remove-punctuation-marks-from-a-string-in-python-3-x-using-translate

```python
punctuation_translator = str.maketrans(' ',' ',string.punctuation)
cleaned_train_comments = train_comments.str.translate(punctuation_translator)
cleaned_test_comments = test_comments.str.translate(punctuation_translator)
```

```python
cleaned_train_comments.head(25)
```

## Removing digits

The Signpost  24 September 2012 Read Signpost --> The Signpost   September  Read Signpost

```python
digit_translator = str.maketrans(' ', ' ', digits)
cleaned_train_comments = cleaned_train_comments.str.translate(digit_translator)
cleaned_test_comments = cleaned_test_comments.str.translate(digit_translator)
```

```python
cleaned_train_comments.head(25)
```

## Removing newline characters

```python
newline_translator = str.maketrans(' ',' ', '\n')
cleaned_train_comments = cleaned_train_comments.str.translate(newline_translator)
cleaned_test_comments = cleaned_test_comments.str.translate(newline_translator)
```

```python
cleaned_train_comments.head(25)
```

## Split combined words

```python
cleaned_train_comments=cleaned_train_comments.apply(lambda comment: re.sub(r'([a-z])([A-Z])',r'\1 \2',comment))
cleaned_test_comments = cleaned_test_comments.apply(lambda comment: re.sub(r'([a-z])([A-Z])', r'\1 \2', comment))
```

```python
cleaned_train_comments.head(25)
```

## Convert to lowercase

```python
cleaned_train_comments = cleaned_train_comments.str.lower()
cleaned_test_comments = cleaned_test_comments.str.lower()
```

```python
cleaned_train_comments.head(25)
```

## Removing stopwords 

```python
cleaned_train_comments.shape
```

```python
# removing stopwords from train data
stop_words = set(stopwords.words("english"))
for i in range(cleaned_train_comments.shape[0]):
    print(i)    
    original_comment = word_tokenize(cleaned_train_comments[i])
    filtered_comment = []
    for word in original_comment:
        if word == "n't":
            word = "not"
        if word not in stop_words:
            filtered_comment.append(word)
        
        
#     filtered_comment = [w for w in original_comment if w not in stop_words]
    train_data["Comment"][i] = ' '.join(w for w in filtered_comment)
```

```python
train_data = train_data.drop(["Unnamed: 0"], axis =1)
train_data.head(25)
```
